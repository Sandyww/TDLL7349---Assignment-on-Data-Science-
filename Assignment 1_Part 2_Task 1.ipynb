```python
import pandas as pd

# Load the dataset
file_path = r'file_path.csv' # the file path of the dataset with missing values
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

## 1. Inspecting the Dataset for Missing Values

Let's check which columns have missing values and how many missing values are present in each column.

```python
# Check for missing values in each column
df.isnull().sum()
```

## 2. Imputing Missing Values

We will use the following strategies:
- For numerical columns: Impute missing values with the column mean.
- For categorical columns: Impute missing values with the column mode (most frequent value).

Let's perform the imputation.

```python
# Impute missing values
for col in df.columns:
    if df[col].dtype == 'object':
        # Fill categorical columns with mode
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Fill numerical columns with mean
        df[col].fillna(df[col].mean(), inplace=True)

# Verify that there are no missing values left
df.isnull().sum()
```

## 3. Saving the Cleaned Dataset

Now that all missing values have been imputed, we will save the cleaned dataset to a new CSV file named `surname_studentID_dataset3_solution.csv`.

Replace `surname` and `studentID` with your actual surname and student ID.

```python
# Save the cleaned dataset
output_path = r'output_path.csv'  # Change surname and studentID

df.to_csv(output_path, index=False)
print(f"Cleaned dataset saved to: {output_path}")
```
