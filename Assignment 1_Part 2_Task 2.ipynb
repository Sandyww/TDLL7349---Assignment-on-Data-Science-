We will use `dataset4_with_outliers.csv` as our example.

```python
# Import the pandas library, which is used for data manipulation
import pandas as pd
# Import the numpy library, which is used for numerical operations
import numpy as np

# Load the dataset
file_path = r'file_path.csv'  # Path to the CSV file

df = pd.read_csv(file_path)  # Read the CSV file into a DataFrame

# Display the first few rows of the DataFrame to see what the data looks like
df.head()
```

## 1. Inspecting the Dataset for Outliers

Let's use the IQR (Interquartile Range) method to detect outliers in each numerical column.

```python
# Identify outliers using the IQR method

def find_outliers_iqr(data, column):
    # Calculate the first quartile (25th percentile)
    Q1 = data[column].quantile(0.25)
    # Calculate the third quartile (75th percentile)
    Q3 = data[column].quantile(0.75)
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    # Calculate the lower bound for outliers
    lower_bound = Q1 - 1.5 * IQR
    # Calculate the upper bound for outliers
    upper_bound = Q3 + 1.5 * IQR
    # Return rows where the value is less than the lower bound or greater than the upper bound (outliers)
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Find outliers for each numerical column
outlier_summary = {}  # Create an empty dictionary to store the number of outliers for each column
for col in df.select_dtypes(include=[np.number]).columns:  # Loop through each numerical column in the dataframe
    outliers = find_outliers_iqr(df, col)  # Find outliers in the current column
    outlier_summary[col] = len(outliers)  # Store the number of outliers in the dictionary

outlier_summary  # Show the summary of outliers for each column
```

```python
# Total number of outlier rows (rows with any outlier in any numerical column)
outlier_rows = set()  # Create an empty set to store the indices of outlier rows
for col in df.select_dtypes(include=[np.number]).columns:  # Loop through each numerical column
    
    # Find the indices of outlier rows in the current column and add them to the set
    outlier_rows.update(find_outliers_iqr(df, col).index)

# Print the total number of unique outlier rows
print(f"Total number of outlier rows: {len(outlier_rows)}")
```

## 2. Removing Outliers

We will remove rows that contain outliers in any numerical column using the IQR method.

```python
# Remove outliers from all numerical columns
def remove_outliers_iqr(data):
    # Loop through each numerical column in the DataFrame
    for col in data.select_dtypes(include=[np.number]).columns:
        # Calculate the first quartile (25th percentile)
        Q1 = data[col].quantile(0.25)
        # Calculate the third quartile (75th percentile)
        Q3 = data[col].quantile(0.75)
        # Calculate the interquartile range (IQR)
        IQR = Q3 - Q1
        # Calculate the lower bound for outliers
        lower_bound = Q1 - 1.5 * IQR
        # Calculate the upper bound for outliers
        upper_bound = Q3 + 1.5 * IQR
        # Keep only the rows where the value is within the bounds (not an outlier)
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data  # Return the cleaned DataFrame

# Apply the function to remove outliers and store the result in a new DataFrame
df_clean = remove_outliers_iqr(df)

# Check for missing values in the cleaned DataFrame
df_clean.isnull().sum()
```

## 3. Saving the Cleaned Dataset

Now we save the cleaned dataset to a new CSV file named `surname_studentID_dataset4_solution.csv`.
